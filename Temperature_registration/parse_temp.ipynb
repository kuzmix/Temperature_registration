```python
import pandas as pd
import datetime
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
folder = Path('../data')  # или Path('.') для текущей
files = [f for f in folder.iterdir() if f.is_file()]
print(files)
```

```python
file =files[0]
data = pd.read_csv(file, encoding="utf-8")
```

```python
data["timestamp"] = pd.to_datetime(data["timestamp"])

wide = data.pivot(
    index="timestamp",        # строки
    columns="channel",        # колонки = каналы
    values="temperature_C",   # значения = температура
)
wide = wide.resample("1s").mean()
```

```python
wide.plot( figsize=(16, 8), sharex=True)
```
